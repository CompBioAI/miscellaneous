```python
# Read location of the file
file_loc = "C:/Users/Kundan/Desktop/kundan/BRCA1_datasets/ncbi_dataset/data/barca1.fasta"
```

```python
# Define a function to get the sequence of the gene
def read_seq(file_loc):
    sequence = ''
    with open(file_loc, "r") as f:
        for line in f:
            if line[0]!='>':
                sequence += line.rstrip()
    return sequence
```

```python
# Print the sequence
read_seq(file_loc)
```

```python
# Define the function to get the base count
def base_count(ab):
    # get the sequence using the above difined function
    sequence = read_seq(ab)
    
    A = 0
    T = 0 
    C = 0
    G = 0
    n = 0
    for base in sequence:
        if base == 'A':
            A = A+1
        elif base == 'T':
            T = T+1
        elif base == 'G':
            G = G+1
        elif base == 'C':
            C = C+1
        else:
            n = n+1
    return [len(sequence), A, T, G, C, n]
```

```python
# get the base count as list
results = base_count(file_loc)
```

```python
# Print the base count and the number of each base present in a gene
print("Total base count = :", results[0],"\n",
      "Total A :", results[1],"\n",
      "Total T :", results[2],"\n",
      "Total G :", results[3],"\n",
      "Total C :", results[4],"\n",
      "Total Unknown :", results[5] 
     )
```
